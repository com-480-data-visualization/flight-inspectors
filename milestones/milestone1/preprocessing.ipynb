```python
import pandas as pd

PATH = "./data/"
```

```python
df_raw = pd.read_csv(PATH+"crashes.csv")
df_raw.head()
```

```python
df = df_raw[["Date", "Time", "Location", "Operator", "Flight #", "Route", "AC Type", "Aboard", "Fatalities", "Ground"]]
df["Date"] = pd.to_datetime(df["Date"], format="%B %d, %Y")
df["Year"] = df["Date"].dt.year
df["Operator"] = df["Operator"].str.split(" /").str[0].str.split(" -").str[0].str.lower().astype(str)
df["Operator"] = df["Operator"].str.replace("air lines", "airlines")
df["Aboard"] = pd.to_numeric(df["Aboard"].str.split().str[0], errors="coerce")
df["Fatalities"] = pd.to_numeric(df["Fatalities"].str.split().str[0], errors="coerce")
df["AC mfr"] = df["AC Type"].str.split().str[0].str.lower().astype(str)
df["Departure"] = df["Route"].str.split(" -").str[0].where(df["Route"].str.contains("-"), "?").astype(str)
df["Arrival"] = df["Route"].str[::-1].str.split(" -").str[0].str[::-1].where(df["Route"].str.contains("-"), "?").astype(str)
df.head()
```

```python
df.to_csv(PATH+"crashes_cleaned.csv")
```
